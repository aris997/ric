# v0.0.0

a very basic differencerator, takes two files, watch the cell (row, column) if
differ from the new (right file), if that's true record the change into a json
like variable. It can be saved in a blob file.


```python
from itertools import zip_longest


with open('hello.txt', 'r') as left, open('hello2.txt', 'r') as right:
    row, column, max_column = 0, 0, 0
    changes = []
    changing = False
    for chleft, chright in zip_longest(left.read(), right.read(), fillvalue=None):
        if chleft != chright:
            if not changing:
                changes.append({
                    'row':row,
                    'col': column,
                    'change': chright})
                changing = True
            else:
                changes[-1]['change'] += chright
        else:
            if changing:
                changing = False
        column += 1
        if chleft is not None and ord(chleft) == 10:
            if column > max_column:
                max_column = column
            column = 0
            row += 1
changes
```

# v0.0.1

look up hashes of rows, if they differ then save the change.

```python
from itertools import zip_longest

with open('hello.txt', 'r') as left, open('hello2.txt', 'r') as right:
    row = 0
    changes = []
    changing = False
    left_rows = left.read().split('\n')
    right_rows = right.read().split('\n')
    last_row = len(max(left_rows, right_rows)) - 1
    for left_row, right_row in zip_longest(left_rows, 
                                           right_rows, 
                                           fillvalue=None):
        if hash(left_row) != hash(right_row):
            if not changing:
                changes.append({
                    'row': row,
                    'change': [right_row]
                })
                changing = True
            else:
                changes[-1]['change'].append(right_row)
                if last_row == row:
                    changes[-1]['change'] = f'{chr(10)}'.join(changes[-1]['change'])
        else:
            if changing:
                changes[-1]['change'] = f'{chr(10)}'.join(changes[-1]['change'])
                changing = False
        row += 1
changes
```
